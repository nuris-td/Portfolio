```python
def fill_not_O_nan_median(data):
    for column in data.keys():
        if (data[column].isna().sum() > 0) & (data[column].dtype!='object'):
            data[column]=data[column].fillna(data[column].median())
def fill_O_nan_mode(data):
    for column in data.keys():
        if (data[column].isna().sum() > 0) & (data[column].dtype=='object'):
            data[column]=data[column].fillna(data[column].mode()[0])
```

```python
fill_not_O_nan_median(df)
fill_O_nan_mode(df)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def Encoder(data):
    for column in data.keys():
        if data[column].dtype=='object':
            col_le = LabelEncoder()
            col_le.fit(data[column])
            data[column]=col_le.transform(data[column])
```
